Upload and Extract Text from Your PDF


```python
from google.colab import files
import fitz  # PyMuPDF

# --- STEP 1: Upload from File Explorer ---
uploaded = files.upload()  # You’ll get a file chooser popup
pdf_file = list(uploaded.keys())[0]

print(f"✅ Uploaded: {pdf_file}")
```

```python
!pip install -q pymupdf

import fitz  # PyMuPDF

def extract_text_pymupdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text("text")
    return text

# --- STEP 2: Extract text ---
raw_text = extract_text_pymupdf(pdf_file)

print("✅ Extracted text successfully!\n")
print(raw_text[:800])  # Preview first 800 characters
```

cleaning the text

```python
import re #This code cell defines a function called clean_text that takes a string of text as input and performs several cleaning steps
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [t for t in tokens if t not in stop_words and len(t) > 2]
    return tokens

clean_tokens = clean_text(raw_text)
print("✅ Cleaned sample tokens:")
print(clean_tokens[:40])
```

frequency visualization

```python
from collections import Counter
#It analyzes the frequency of the cleaned tokens and
#visualizes the top 20 most frequent words.


freq = Counter(clean_tokens)
common = freq.most_common(20)
words , counts = zip(*common)
plt.figure(figsize =(10,5))
plt.bar(words , counts)
plt.xticks(rotation = 45)
plt.title("top 20 Frequent Words")
plt.show()
```

save preprocessed tokens

```python
import json
with open ("cleaned_tokens.json", "w") as f:
  json.dump(clean_tokens, f)
  print("cleaned tokens saved")
  files.download("cleaned_tokens.json")
```
